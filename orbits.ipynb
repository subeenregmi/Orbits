```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np 
import math
from IPython import display
%matplotlib inline
```

```python
GRAVITY_CONSTANT = -5
```

```python
class Planet:
    def __init__(self, mass, initial_velocity, position):
        self.mass = mass
        self.velocity = initial_velocity
        self.position = position

    def calculateVelocity(self, planets, tickSpeed):
        for p in planets:
            distance = self.position - p.position
            if not np.any(distance):
                continue
            force = (self.position - p.position) * GRAVITY_CONSTANT * p.mass * 1/((math.sqrt(sum(pow(element, 2) for element in (distance))))**3)
            self.velocity = self.velocity + (force * tickSpeed)

    def calculatePosition(self, tickSpeed):
        self.position = self.position + (self.velocity * tickSpeed)
```

```python
def simulate(Planets, frames=1000, tickspeed=0.1):
    fig = plt.figure()
    lines = plt.plot([], 'o')
    
    line = lines[0]
    
    plt.xlim(-2000, 2000)
    plt.ylim(-2000, 2000)
    
    def animate(frame):
        xs = []
        ys = []
        for p in Planets:
            p.calculateVelocity(Planets, 0.1)
        for p in Planets:
            p.calculatePosition(0.1)
            xs.append(p.position[0])
            ys.append(p.position[1])
        
        line.set_data(xs, ys)
    
        
    anim = FuncAnimation(fig, animate, frames=785, interval=10)
    video = anim.to_html5_video()
    html = display.HTML(video)
    display.display(html)
    plt.close()
```

```python
p1 = Planet(1, np.array([-75, 75]), np.array([-1000, -1000]))
p2 = Planet(3300000, np.array([0, 0]), np.array([0, 0]))
simulate([p1, p2])
```
